# Perceptron prosty

## Algorytm uczenia perceptronu prostego

Zakładamy, że klasy $y_i \in \{-1, +1\}$<br>
Wartości początkowe: $N=100$ ilość iteracji, $\eta = 1$ wspólczynnik uczenia

1. Zainicjuj wagi $\mathbf{w}$ i wyraz wolny $b$ wartościami 0
2. Powtarzaj $N$ razy
3. <ul> wybierz wektor $\mathbf{x}_i$ z klasy $y_i$ </ul> 
3. <ul> oblicz aktywację <br> $z = \text{sign}(\mathbf{x}_i\mathbf{w} + b)$</ul>
3. <ul> aktualizacja wag <br> 
$
\begin{aligned}
& \mathbf{w} \leftarrow \mathbf{w}+\eta\left(y_i-z\right) \mathbf{x}_i \\
& b \leftarrow b+\eta\left(y_i-z\right)
\end{aligned}
$</ul>


## Klasa perceptron - klasyfikator

```python
import numpy as np
import random

class Perceptron(object):
    """Perceptron - klasyfikator binarny

    Parametry
    ------------
    eta : współczynnik uczenia (w przedziale pomiędzy 0.0 a 1.0)
    n_iter : liczba iteracji

    Atrybuty
    -----------
    w : wektor wag
    b : wyraz wolny (bias)

    """
    def __init__(self, eta=1.0, n_iter=100):
        self.eta = eta
        self.n_iter = n_iter
     
    def predict(self, X):
        """Zwraca etykietę klas po obliczeniu funkcji skoku jednostkowego"""
        return np.sign(np.dot(X, self.w) + self.b)
    
    def error(self, X, y):
        return (self.predict(X) != y).sum()
    
    def fit(self, X, y):
        """Dopasowanie danych uczących.

        Parametry
        ----------
        X : wektory uczące, wymiary = [n_próbek, n_cech]
        y : wartości docelowe, wymiary = [n_próbek]

        Zwraca
        -------
        self : obiekt

        """
        
        
        return self   
```

## Przygotowanie danych teningowych

* wybierzmy pierwsze 100 przypadków, dwie klasy: `Setosa`, `Versicolor`
* wybierzmy dwie zmienne: `sepallenght`  oraz `petallenght`

```python
import numpy as np

# wybiera gatunki setosa i versicolor
y = df.iloc[0:100, 4].values
y = np.where(y == 'Iris-setosa', -1, 1)

# wydobywa cechy: długość działki i długość płatka
X = df.iloc[0:100, [0, 2]].values
```

Wykres rozrzutu

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(X[:50, 0], X[:50, 1], color='red', marker='o', label='Setosa')
plt.scatter(X[50:100, 0], X[50:100, 1], color='blue', marker='x', label='Versicolor')
plt.xlabel('Długość działki [cm]')
plt.ylabel('Długość płatka [cm]')
plt.legend(loc='upper left')
plt.show()
```

## Uczenie perceptronu - dane separowalne liniowo


```python
ppn = Perceptron(eta=1, n_iter=1000)
ppn.fit(X, y)

print('errors=', ppn.error(X, y), ', w=', ppn.w, ', b=', ppn.b)

plt.figure(figsize=(20,5))
plt.plot(range(1, len(ppn.errors) + 1), ppn.errors, marker='o')
plt.xlabel('Iteracje')
plt.ylabel('Liczba błędów')
plt.show()
```

## Granica decyzyjna

Granica decyzyjna jest zdefiniowana przez wyrażenie

$$ w_0 x_0 + w_1 x_1 + b = 0$$

stąd ptrzymujemy linie prostą 
$$x_1=-\frac{w_0}{w_1} x_0+\frac{b}{w_1}$$




```python
a = -ppn.w[0]/ppn.w[1]
b = ppn.b/ppn.w[1]

x1 = np.linspace(X[:,0].min(), X[:,0].max(), 50)
y1 = a * x1+ b

plt.scatter(X[:50, 0], X[:50, 1], color='red', marker='o', label='Setosa')
plt.scatter(X[50:100, 0], X[50:100, 1], color='blue', marker='x', label='Versicolor')
plt.plot(x1, y1)
plt.show()
```

## Funkcja rysująca granicę decyzyjną klasyfikatora

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=0.02):

    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                         np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # rysuje wykres próbek
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1], alpha=0.8, c=colors[idx], marker=markers[idx], label=cl)
```

```python
plot_decision_regions(X, y, classifier=ppn)
plt.xlabel('Długość działki [cm]')
plt.ylabel('Długość płatka [cm]')
plt.legend(loc='upper left')

plt.show()
```

## Funkcja pomocnicza: raport

```python
def raport(perceptron, X, y):
    print('w        = ', perceptron.w)
    print('b        = ', perceptron.b)
    print('errors   = ', perceptron.error(X,y))
    print('accuracy = ', 1.0-perceptron.error(X,y)/X.shape[0])
    
    plt.figure(figsize=(20,5))
    plt.plot(range(1, len(ppn.errors) + 1), ppn.errors, marker='o')
    plt.xlabel('Iteracje')
    plt.ylabel('Liczba błędów')
    plt.show()
    
    plot_decision_regions(X, y, classifier=ppn)
    plt.xlabel('Długość działki [cm]')
    plt.ylabel('Długość płatka [cm]')
    plt.legend(loc='upper left')
    plt.show()

ppn = Perceptron(eta=1, n_iter=1000).fit(X,y)
raport(ppn, X, y)
```

## Pseudo-losowanie permutacyjne

* **Epoka** - okres uczenia, w którym kazdy wektor treningowy jest prezentowany jeden raz
* Przed wykonaniem kolejnej epoki dokonujemy wymieszania wektorów - każda epoka posiada inną kolejność prezentacji
* Każdy wektor ma sprawiedliwy wkład w proces uczenia

---
Algorytm

1. Ustaw ilość epok $N$ oraz wspólczynnik uczenia $\eta$
1. Zainicjuj wagi $\mathbf{w}$ i wyraz wolny $b$
1. Powtarzaj $N$ razy
1. <ul> wymieszaj kolejność wektorów treningowych</ul>
1. <ul> dla każdego $i=0,1,2,\ldots, n$</ul>
1. <ul><ul>wykonaj krok uczenia<br>
$\begin{aligned}
& \mathbf{w} \leftarrow \mathbf{w}+\eta\left(y_i-z\right) \mathbf{x}_i \\
& b \leftarrow b+\eta\left(y_i-z\right)
\end{aligned}$</ul></ul>

## Algorytm kieszeniowy

Wartości początkowe: $N=100$ ilość iteracji, $\eta = 1$ wspólczynnik uczenia

1. Zainicjuj wagi $\mathbf{w}$ i wyraz wolny $b$, czas $t=0$
1. Zapamiętaj wagi w kieszonce $\quad t^{\prime} \leftarrow t, \quad \mathbf{w}^{\prime} \leftarrow \mathbf{w}, \quad b^{\prime} \leftarrow b$
1. Powtarzaj $N$ razy
1. <ul> wybierz wektor $\mathbf{x}_i$ należący do klasy $y_i$ </ul> 
1. <ul> oblicz aktywację  $z = \text{sign}(\mathbf{x}_i\mathbf{w} + b)$</ul>
1. <ul> jeśli $z = y_i$ <br>
1. <ul><ul> $t \leftarrow t + 1$ </ul></ul>
1. <ul> w przeciwnym wypadku<br>
1. <ul><ul> jeśli $t > t'$ wykonaj</ul></ul>
1. <ul><ul><ul>$t' \leftarrow t, \quad \mathbf{w}' \leftarrow \mathbf{w}, \quad b' \leftarrow b$</ul></ul></ul>
1. <ul><ul>$t = 0$<br>
$\begin{aligned}
& \mathbf{w} \leftarrow \mathbf{w}+\eta\left(y_i-z\right) \mathbf{x}_i \\
& b \leftarrow b+\eta\left(y_i-z\right)
\end{aligned}
$</ul></ul>
1. Zwróć wektor zapamiętany w kieszonce $\mathbf{w}', b$
